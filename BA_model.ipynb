```python
# Compute clustering coefficients
clustering_values = list(nx.clustering(BA_network).values())

# Plot clustering coefficient distribution
plt.figure(figsize=(8, 5))
plt.hist(clustering_values, bins=30, color="skyblue", edgecolor="black", alpha=0.7)
plt.xlabel("Clustering Coefficient")
plt.ylabel("Number of Nodes")
plt.title("Clustering Coefficient Distribution Barab√°si-Albert")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
# Compute betweenness centrality
betweenness_values = list(nx.betweenness_centrality(BA_network).values())

# Plot betweenness centrality distribution
plt.figure(figsize=(8, 5))
plt.hist(betweenness_values, bins=30, color="lightcoral", edgecolor="black", alpha=0.7)
plt.xlabel("Betweenness Centrality")
plt.ylabel("Number of Nodes")
plt.title("Betweenness Centrality Distribution Barab√°si-Albert")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
# Find connected components sizes
component_sizes = [len(c) for c in nx.connected_components(BA_network)]

# Plot component size distribution
plt.figure(figsize=(8, 5))
plt.hist(component_sizes, bins=10, color="orange", edgecolor="black", alpha=0.7)
plt.xlabel("Component Size")
plt.ylabel("Number of Components")
plt.title("Connected Components Size Distribution Barab√°si-Albert")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
